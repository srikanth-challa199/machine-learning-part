```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import joblib
import matplotlib.pyplot as plt

df=pd.read_csv('/content/polydataset(polynomial regression).csv')

x=df['Level'].values.reshape(-1, 1)

y=df['Salary']

plt.scatter(x,y,color='blue')

poly=PolynomialFeatures(degree=4)
variable=poly.fit_transform(x)

model=LinearRegression()
model.fit(variable,y)


plt.plot(x,model.predict(variable),color='black')
plt.show()


new_input_transformed = poly.fit_transform([[3.5]])
print(model.predict(new_input_transformed))


joblib.dump(model, 'model_joblib')


vega=joblib.load('model_joblib')


theta=PolynomialFeatures(degree=4)


vega = theta.fit_transform([[3.5]])


print(model.predict(vega))
```
