```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv('/content/homeprices2.csv')
print(df)

dummies=pd.get_dummies(df['town']).astype(int) # Corrected syntax: .astype(int) is a method call
con=pd.concat([df.drop('town', axis='columns'),dummies], axis='columns') # Drop original 'town' and concatenate
print(con)

X=con.drop('price',axis='columns')
y=con['price']

model=LinearRegression()
model.fit(X,y)

model.predict([[3800,0,1,0]])
```
