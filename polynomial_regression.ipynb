```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

df=pd.read_csv('/content/polydataset(polynomial regression).csv')
print(df)

x=df['Level'].values.reshape(-1, 1)

y=df['Salary'].values

plt.scatter(x,y,color='blue')

poly=PolynomialFeatures(degree=6)
variable=poly.fit_transform(x)

model=LinearRegression()
model.fit(variable,y)

plt.plot(x,model.predict(variable),color='red')
plt.show()

print(model.predict(poly.fit_transform([[4.5]])))
```
