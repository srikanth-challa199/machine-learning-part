```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

df=pd.read_csv('income.csv')
print(df)

km=KMeans(n_clusters=3, random_state=0, n_init=10)
scaler=MinMaxScaler()
new=scaler.fit_transform(df[['Age']])
df['Age']=new
print(df)
old=scaler.fit_transform(df[['Income($)']])
df['Income($)']=old
print(df)

v=km.fit_predict(df[['Age','Income($)']])
df['cluster']=v
print(df)

df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='red',marker='*')
plt.scatter(df2['Age'],df2['Income($)'],color='blue',marker='*')
plt.scatter(df3['Age'],df3['Income($)'],color='brown', marker='*')
plt.show()

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black')

print(km.cluster_centers_)
```
