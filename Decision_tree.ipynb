```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv('/content/salaries(decision tree).csv')
encoder=LabelEncoder()
X=df.drop('salary_more_then_100k',axis='columns')
y=df['salary_more_then_100k']
print()
X['companies_n']=encoder.fit_transform(df['company'])
X['job_n']=encoder.fit_transform(df['job'])
X['degree_n']=encoder.fit_transform(df['degree'])
print(X)
print(y)
final=X.drop(['company','job','degree'],axis='columns')
print(final)
print(y)
model=DecisionTreeClassifier()
model.fit(final,y)
print(model.predict([[5,2,9]]))
```
