```python
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import numpy as np

digits=load_digits()

a=LogisticRegression(solver='lbfgs', max_iter=5000)

b=SVC()

c=RandomForestClassifier(n_estimators=50)

Score1=cross_val_score(a,digits.data,digits.target,cv=3)
Score2=cross_val_score(b,digits.data,digits.target,cv=3)
Score3=cross_val_score(c,digits.data,digits.target,cv=3)

print(np.average(Score1))
print(np.average(Score2))
print(np.average(Score3))
```
