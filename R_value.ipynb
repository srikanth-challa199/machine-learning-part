```python
import pandas as pd
from scipy import stats

d = {"a":[600, 3000, 2, 3600, 4],"b":[550000, 565000, 610000, 680000, 725000]}

df=pd.DataFrame(d)
print(df)
print()
X=df['a']
y=df['b']

print(X)
print()
print(y)

slope,intercept,r_value,p_value, stud_err=stats.linregress(X,y)

print(r_value)
```
